customising graphs -> label,legends , linewidth, axis , grid , 

```python
#plt.plot(x,y,color="black",marker="+"")
y2=x**2
plt.plot(x,y,color="black",linewidth=5,label="x ^3")
plt.plot(x,y2,color="red",linewidth=4,label="x ^2")
plt.legend() # ti display line wise label
#plt.label for overall label also can be used
plt.axis([0,10,1,100]) # x from 1-->10 , y from 1-->100
plt.ylabel("x ^3")
plt.xlabel("x")
plt.grid() # to plot grid over G

# to show text at a lovation in G
plt.text(2,80,"text",fontsize=12,color="blue") # wher to show text
plt.title("x power 3")
plt.show()
```

Bubble charts  -> (advantage over scatter plot) the size of point can signify information<br/>
eg : can give 3 info at a time <br/>
x : year , Y : avg salary , size of point : No of people 

```python
year=[2012,2013,2014,2015,2016,2017]
sal=[12,13,14,17,19,12]
population=[100,120,200,220,300,390]
```

```python
plt.scatter(year,sal,s=100) # size if 100
plt.show()
c=np.arange(len(year)) # used to give colors
plt.scatter(year,sal, s=population, c=c,alpha=0.6 , edgecolor='b' ,marker="o") # can give size as an array
#alpha-->transparency
#edgecolor for the bubble 
# marker also

# for loop for text
for i in range(len(year)):
    plt.text(year[i]+0.1,sal[i]+0.1,"pop"+"{:d}".format(i))
plt.show()

# array for color how it wokrs --> if 2 index
```

```python
import matplotlib.pyplot as plt
import numpy as np
#PIE GRAPHS
###############
labels=["A","B","C","D"]
sizes=[3,4,6,2]
colors=["blue","red","yellow","purple"]
plt.pie(sizes,colors=colors,labels=labels,autopct="%.1f %%") # after decimal show one point in autopct
plt.title("Pie chart")
plt.axis("equal")
plt.show()
# for autopct can either pass fucntion or just pass a formatted string
##he autopct is used to format value of each label. Pie chart is used to show parts and % 
# to show % symbol add 2 %% at end 


```

```python
## TO EXplode --> to highlight one slics 
## value 1 --> completely out of pie 

explode=[0.1,0.0,0.0,0.0]
# explode to hihglight one pie
plt.pie(sizes,colors=colors,explode=explode ,labels=labels,autopct="%.1f %%",counterclock=False, startangle=100)
 # counterclock to False (deafult is True)
    #startsnagle can be specified
    
```

```python
plt.hist(a,edgecolor="black")
plt.xticks([1,2,3,4,5])
plt.show()
```

```python
xt=np.arange(23)
plt.hist(a,edgecolor="black")
plt.xticks(xt)
plt.show()
```

The value of ranges look bit wierd like 3.somthin to 5.2 etc

and 10 such bit are made default 


```python
# to sepecify bins
plt.hist(a,bins=25,edgecolor="black") # smaller towers are formed since now 25 earlier only 10 bins
plt.xticks(xt)
plt.show()
# din size of 22 might extacly give width one for our data
#the din of form [a, b)  ..expcet last bin [x,y] inclusive of both end 
```

```python
plt.hist(a,bins=[1,3,6,10,15,22],edgecolor="black")  # bins decided by us 
plt.xticks(xt)
plt.axis([0,25,0,10]) # x min, x max , y min, y max (axis size chnages )
plt.show()
# bins are [1,3) , [3,6) ..so on
```
